   <li><b>2D and 3D plotting:</b> Can generate both 2D and 3D plots.</li>

 <li><b>Multiple output formats:</b> Matplotlib can save figures in many formats such as PNG, JPG, SVG, and PDF.</li>

```python
import seaborn as sb
import matplotlib.pyplot as plt
iris=sb.load_dataset('iris')
print(iris.head())
```

<h1> 1. General Statistics Plot (Matplotlib or Seaborn)</h1>

The goal of this task is to give a general statistical summary of the
Iris dataset. This can be done using:

<h3> Seaborn’s pairplot:</h3>

It visualizes pairwise relationships between variables (like sepal
length, sepal width, petal length, petal width), showing scatter plots
and histograms.

<h3>Pandas’ describe() function:</h3>

It provides a statistical summary of the data, including count, mean,
min, max, standard deviation, and quartiles for each numeric feature.

<h4>Purpose:</h4>

To understand the overall distribution and relationships among features
of the Iris dataset.

```python
 #General statistical summary using seaborn's pairplot
sb.pairplot(iris, hue='species')
plt.show()
# Alternatively, to use Pandas' describe()
 print(iris.describe())
```

<h2>Pie Plot for Species Frequency:</h2>

This task is to create a pie chart to display the frequency of the three
species in the Iris dataset: Setosa, Versicolor, and Virginica.

Pie charts are circular graphs divided into sectors, where each sector
represents a proportion of the whole dataset. It helps show the
proportion of each species in the dataset.

<h4>Purpose:</h4>

To easily visualize how the three species are distributed in the
dataset.

```python
 # Pie chart for species frequency
species_counts = iris['species'].value_counts()
plt.figure(figsize=(6,6))
# pie chart
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Species Frequency in Iris Dataset')
plt.show()
```

<h1> Relationship Between Sepal Length and Width:</h1>

Relationship Between Sepal Length and Sepal Width

This task involves creating a scatter plot to find the relationship
between sepal length and sepal width.

Scatter plots are used to show the relationship or correlation between
two variables. Each point represents a data observation.

<h4>Purpose:</h4>

To determine if there is a linear, non-linear, or no relationship
between sepal length and width.


```python
# Scatter plot to find the relationship between Sepal Length and Sepal Width
plt.figure(figsize=(8,6))
sb.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

<h1>Distribution of Sepal and Petal Features:</h1>

Histograms or KDE plots (Kernel Density Estimate) can be used to

how each feature is spread and whether they are skewed or normally
distributed.

To understand how each feature (sepal and petal measurements) is distributed across the dataset.

<h1>Jointplot of Sepal Length vs Sepal Width:</h1>

A joint plot is a combination of a scatter plot with the individual
distributions of sepal length and sepal width displayed on the same
plot.

The scatter plot shows the relationship between two variables, while
histograms (or KDE plots) along the axes show the individual
distributions of these variables.

<h4>Purpose:</h4>

To analyze the relationship between the two variables along with their respective distributions.

```python
sb.jointplot(x='sepal_length', y='sepal_width', data=iris, hue='species', kind='scatter')
    plt.show()
```

<h4>KDE Plot for Setosa Species (Sepal Length vs Sepal Width):</h4>

This task involves creating a KDE plot (Kernel Density Estimate) for
sepal length versus sepal width for the Setosa species.

KDE plots are a smoothed approximation of a histogram. They help
visualize the probability density of the variables.

<h4>Purpose:</h4>

To see how the Setosa species' sepal measurements are distributed and
where the highest density of observations occurs.

```python
 setosa = iris[iris['species'] == 'setosa']

# KDE plot: Sepal Length vs Sepal Width for Setosa
sb.kdeplot(x='sepal_length', y='sepal_width', data=setosa, cmap='Blues', fill=True, thresh=0.05)
plt.title('KDE Plot: Sepal Length vs Sepal Width for Setosa')
plt.show()
```

<h1>KDE Plot: Petal Length vs Petal Width for Setosa</h1>

Similarly, this task requires a KDE plot for petal length versus petal
width for the Setosa species.

It will show the density distribution of petal measurements for the
Setosa species.

<h4> Purpose:</h4>

To visualize how the petal length and width are distributed in Setosa
species and identify patterns.


```python
 # KDE plot: Petal Length vs Petal Width for Setosa
sb.kdeplot(x='petal_length', y='petal_width', data=setosa, cmap='Purples', fill=True, thresh=0.05)
plt.title('KDE Plot: Petal Length vs Petal Width for Setosa')
plt.show()
```
